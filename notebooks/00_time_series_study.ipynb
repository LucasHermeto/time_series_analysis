#### Random Sequence

Considering $\{X_n, n=1,2,...\}$ a sequence of random variables in the sample space $\Omega$.

For all $n\geq1$, we can write

$$P(X_1=a_1,...,X_n=a_n) = P(X_1=a_1)P(X_2=a_2|X_1=a_1)...P(X_n=a_n|X_1=a_1,...,X_{n-1}=a_{n-1})$$

```python
# Investment application considering the random variable R
total_years = 20
initial_investment = 100
all_returns = [initial_investment]
years_count = [0]
for year in range(0, total_years):
    annual_interest = np.random.uniform(0.04, 0.05, 1)[0]
    return_amount = initial_investment * (1 + annual_interest) ** year
    all_returns.append(return_amount)
    years_count.append(year)

plt.figure(figsize=(15, 6))
plt.bar(years_count, all_returns)
plt.title('Investment Returns Over Years')
plt.ylabel('Returns')
plt.xlabel('Years')
plt.show()
```

#### Markov Chain

Probability of an event occurring, given a previous event.

```python
# Markov chain representation
markov_chain = [
    [0.3, 0.7],
    [0.4, 0.6]
]
pd.DataFrame(data=markov_chain, columns=['E', 'A'], index=['E', 'A'])
```

### Time Series Properties

- **Trend (T)**: indicates the behavior of the data over time, whether it is **increasing**, **decreasing**, or **stable**. Additionally, the trend also denotes the speed of these changes.

- **Cycles (C)**: represent **smooth and repetitive oscillations of rise and fall in time series** data, occurring alongside the trend component. Cyclical movements tend to be irregular.

- **Seasonality (S)**: denotes **periodic oscillations of rise and fall that consistently happen during a specific time of the year, month, week, or day**. These movements are easily predictable and occur at **regular time intervals**.

- **Random Noise ($\epsilon$)**: or error at time **t** refers to irregular variations or unexplained fluctuations, stemming from unexpected and chance events.

r = \frac{\sum_{i=1}^N{(x_i - \overline{x})(y_i - \overline{y})}}{\sqrt{[\sum_{i=1}^n{(x_i - \overline{x})^2}][\sum_{i=1}^n{(y_i - \overline{y})^2}]}}
